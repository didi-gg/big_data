```python
# conexión Local a base en Docker
host = "localhost"
port = '5432'
user = 'user'
password = 'password'
database = 'postgres' #verificar la existencia de la base datos
```

```python
import pandas as pd
import numpy as np
from sqlalchemy.sql import text
from sqlalchemy import create_engine
```

```python
#conexión a la base de datos local (contenedor de docker)
connDockerLocal = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}', echo=False)
```

```python
# Librerías requeridas para la conexión y muestra de información
import psycopg2 as ps
import pandas as pd
```

```python
# Prueba de conexión con host local
try:
    connps = ps.connect(host="localhost",database="postgres",user='user',password='password',port='5432')
except ps.OperationalError as e:
    raise e
else:
    print('Connected!')
```

```python
# Busqueda de verificación de conexión local
query = """
    SELECT * FROM listings_summary
"""

with connDockerLocal.connect().execution_options(autocommit=True) as conn:
    query = conn.execute(text(query))         
df = pd.DataFrame(query.fetchall())
df.head()
```
